**10)Create a new column called bmi category if that bmi is less than 18.5 give it as under weight & if it is between 18.5 to 24.9 give it as normal & if it is between 25 to 30 give as overweight & if it is greater than 30 give it as obesity.**


```python
def bmi_category(bmi):
    if bmi < 18.5:
        return "Underweight"
    elif 18.5 <= bmi <= 24.9:
        return "Normal"
    elif 25 <= bmi <= 30:
        return "Overweight"
    else:
        return "Obesity"

medical['BMI_Category'] = medical['BMI'].apply(bmi_category)
```
